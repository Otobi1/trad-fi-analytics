```python
# Function to get the min and max dates for each ticker
def get_min_max_dates(ticker):
    try:
        # Fetch the historical data for the ticker (using a long period to ensure you capture all available data)
        data = yf.Ticker(ticker).history(period="max")
        
        # Get the minimum and maximum dates if data exists
        min_date = data.index.min().date() if not data.empty else None
        max_date = data.index.max().date() if not data.empty else None
        
        return min_date, max_date
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return None, None

# Loop over tickers and store results
min_max_dates = {}
for ticker in tickers:
    min_date, max_date = get_min_max_dates(ticker)
    min_max_dates[ticker] = {"Min Date": min_date, "Max Date": max_date}

# Convert the results to a DataFrame for better readability
df_min_max = pd.DataFrame(min_max_dates).T
print(df_min_max)
```

```python

import json

# List of tickers (you can use tickers from a CSV or predefined list)
tickers = ["AAPL", "MSFT", "GOOG"]

# Function to get all available data for a ticker
def get_ticker_info(ticker):
    stock = yf.Ticker(ticker)
    
    data = {
        "basic_info": stock.info,  # Basic company info
        "history": stock.history(period="max").to_dict(),  # Full historical data
        "dividends": stock.dividends.to_dict(),  # Dividend history
        "splits": stock.splits.to_dict(),  # Stock split history
        "financials": stock.financials.to_dict(),  # Annual financials
        "quarterly_financials": stock.quarterly_financials.to_dict(),  # Quarterly financials
        "earnings": stock.earnings.to_dict(),  # Annual earnings
        "quarterly_earnings": stock.quarterly_earnings.to_dict(),  # Quarterly earnings
        "sustainability": stock.sustainability.to_dict(),  # ESG scores (if available)
        "recommendations": stock.recommendations.to_dict(),  # Analyst recommendations
        "balance_sheet": stock.balance_sheet.to_dict(),  # Balance sheet data
        "cashflow": stock.cashflow.to_dict(),  # Cash flow data
        "major_holders": stock.major_holders.to_dict(),  # Major stockholders
        "institutional_holders": stock.institutional_holders.to_dict(),  # Institutional holders
        "mutual_fund_holders": stock.mutualfund_holders.to_dict(),  # Mutual fund holders
        "calendar": stock.calendar.to_dict(),  # Upcoming events like earnings
        "options_dates": stock.options,  # Option expiration dates
    }
    
    return data

# Loop over all tickers and store their data
all_ticker_data = {}
for ticker in tickers:
    try:
        all_ticker_data[ticker] = get_ticker_info(ticker)
        print(f"Data fetched for {ticker}")
    except Exception as e:
        print(f"Failed to fetch data for {ticker}: {e}")

# Save the data as JSON for structured storage
with open('ticker_data.json', 'w') as f:
    json.dump(all_ticker_data, f)

# Optional: Save individual sections to CSV (e.g., historical data)
# This is for historical price data only
for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        stock.history(period="max").to_csv(f"{ticker}_history.csv")
        print(f"Historical data saved for {ticker}")
    except Exception as e:
        print(f"Failed to save historical data for {ticker}: {e}")
```

```python
from datetime import datetime, timedelta

# Define the ticker symbol
ticker = "AAPL"

# Load the existing data from the Parquet file
try:
    existing_data = pd.read_parquet(f"{ticker}_history.parquet")
    last_date = existing_data.index[-1]  # Get the last available date in the existing data
except FileNotFoundError:
    print("Historical data file not found. Running initial full data fetch...")
    last_date = None

# Fetch data from the last available date + 1 day to today
start_date = (last_date + timedelta(days=1)).strftime('%Y-%m-%d') if last_date else '1900-01-01'
end_date = datetime.now().strftime('%Y-%m-%d')

# Fetch the new data
stock = yf.Ticker(ticker)
new_data = stock.history(start=start_date, end=end_date)

# If there's new data, append it to the existing data and save it
if not new_data.empty:
    # Append the new data to the existing data
    updated_data = pd.concat([existing_data, new_data])

    # Save the updated data back to the Parquet file
    updated_data.to_parquet(f"{ticker}_history.parquet")
    print(f"Data updated for {ticker}. Latest date: {updated_data.index[-1]}")
else:
    print(f"No new data for {ticker}. Latest available data: {last_date}")
```

```python
import yfinance as yf
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# Define the ticker symbol
ticker = "AAPL"

# Fetch full historical data
stock = yf.Ticker(ticker)
history = stock.history(period="max")

# Add additional columns for partitioning
history['Year'] = history.index.year
history['Month'] = history.index.month
history['Day'] = history.index.day
history['Ticker'] = ticker
history['Sector'] = stock.info.get('sector', 'Unknown')  # Assuming sector is available

# Save data partitioned by Year, Month, Ticker
table = pa.Table.from_pandas(history)

# Write to Parquet with partitioning
pq.write_to_dataset(
    table,
    root_path='historical_data',  # Directory to save Parquet files
    partition_cols=['Year', 'Month', 'Ticker']
)

print("Data saved with partitioning by Year, Month, and Ticker.")
```

```python
import yfinance as yf
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
from datetime import datetime, timedelta
import os

# Define the ticker symbol
ticker = "AAPL"
parquet_dir = 'historical_data'  # Directory where the partitioned Parquet files are stored

# Function to find the last partition by year and month
def get_last_partition(parquet_dir, ticker):
    last_year = None
    last_month = None
    
    # List the partitions for the ticker (organized by Year, Month)
    ticker_path = os.path.join(parquet_dir, f'Ticker={ticker}')
    
    if not os.path.exists(ticker_path):
        print(f"No data found for ticker: {ticker}")
        return None, None
    
    # Find the most recent year
    years = [int(year_dir.split('=')[-1]) for year_dir in os.listdir(ticker_path) if year_dir.startswith('Year=')]
    if years:
        last_year = max(years)
        # Find the most recent month for that year
        year_path = os.path.join(ticker_path, f'Year={last_year}')
        months = [int(month_dir.split('=')[-1]) for month_dir in os.listdir(year_path) if month_dir.startswith('Month=')]
        if months:
            last_month = max(months)
    
    return last_year, last_month

# Function to read the last partition (most recent data)
def get_last_date_from_partition(parquet_dir, ticker, last_year, last_month):
    if last_year is None or last_month is None:
        return None
    
    # Path to the most recent partition
    partition_path = os.path.join(parquet_dir, f'Ticker={ticker}', f'Year={last_year}', f'Month={last_month}')
    
    # Load the Parquet files in that partition
    try:
        partition_data = pd.read_parquet(partition_path)
        last_date = partition_data.index.max()  # Get the last date in the partition
        return last_date
    except Exception as e:
        print(f"Error reading partition: {e}")
        return None

# Get the last partition for the given ticker
last_year, last_month = get_last_partition(parquet_dir, ticker)

# Get the last available date from the most recent partition
last_date = get_last_date_from_partition(parquet_dir, ticker, last_year, last_month)

# Fetch data from the day after the last available date, or from the start if no data exists
if last_date:
    start_date = (last_date + timedelta(days=1)).strftime('%Y-%m-%d')
else:
    start_date = '1900-01-01'  # Start from the beginning if no data exists
end_date = datetime.now().strftime('%Y-%m-%d')

# Fetch the new data
stock = yf.Ticker(ticker)
new_data = stock.history(start=start_date, end=end_date)

# If new data is available, partition it and append
if not new_data.empty:
    # Add partitioning columns
    new_data['Year'] = new_data.index.year
    new_data['Month'] = new_data.index.month
    new_data['Day'] = new_data.index.day
    new_data['Ticker'] = ticker
    new_data['Sector'] = stock.info.get('sector', 'Unknown')

    # Convert to pyarrow table
    table = pa.Table.from_pandas(new_data)

    # Append the new data to the Parquet dataset (partitioned by Year, Month, Ticker)
    pq.write_to_dataset(
        table,
        root_path=parquet_dir,
        partition_cols=['Year', 'Month', 'Ticker']
    )
    print(f"Data updated for {ticker} and appended with partitioning.")
else:
    print(f"No new data for {ticker} since {last_date}.")
```
